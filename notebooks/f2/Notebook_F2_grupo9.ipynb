##**5. Dataset**

El dataset cuenta con 8 variables en total. Cada variable representa lo siguiente:

1) UTC Date: Fecha y hora del evento sísmico, expresada en el Tiempo Universal Coordinado (UTC). Es la referencia estándar utilizada globalmente en sismología.

2) Profoundity (Profundidad): Distancia vertical (medida en kilómetros) desde la superficie de la Tierra hasta el hipocentro o foco exacto donde se originó la ruptura del sismo.

3) Magnitude (Magnitud): Medida cuantitativa de la energía liberada en el foco del sismo. Incluye el valor numérico y el tipo de escala empleada.

4) Date (Fecha local): La fecha calendario en la que se registró el sismo ajustada al huso horario local de Chile.

5) Hour (Hora local): La hora exacta del registro del sismo ajustada al huso horario local de Chile.

6) Location (Ubicación): Descripción geográfica o referencia textual que indica a cuántos kilómetros y en qué dirección cardinal se localizó el epicentro respecto a una comuna, ciudad o localidad de referencia.

7) Latitude (Latitud): Coordenada geográfica que indica la posición del epicentro al norte o al sur del ecuador terrestre (expresada en grados decimales, con valores negativos para el hemisferio sur).

8) Longitude (Longitud): Coordenada geográfica que determina la posición del epicentro al este u oeste del meridiano de Greenwich (expresada en grados decimales, con valores negativos para el hemisferio occidental).

###**8.2) Limpieza de datos**

###**8.3) Conversión de columnas de fecha a formato datetime de variable Fecha UTC**



Esta función se genera una columna independiente para las variables año, mes y día de la ocurrencia de terremotos para el formato UTC.

```python
#Descomposición de fecha UTC
def procesar_fecha_utc(df):
    df = df.copy()
    df['FECHA_UTC'] = pd.to_datetime(df['FECHA_UTC'], errors='coerce')
    df['ANIO_UTC'] = df['FECHA_UTC'].dt.year
    df['MES_UTC'] = df['FECHA_UTC'].dt.month
    df['DIA_UTC'] = df['FECHA_UTC'].dt.day
    df['HORA_UTC'] = df['FECHA_UTC'].dt.strftime('%H:%M:%S')

    # Displaying relevant columns that this function modifies
    print(df[['FECHA_UTC', 'ANIO_UTC', 'MES_UTC', 'DIA_UTC', 'HORA_UTC']].head())
    return df
```

###**8.4) Separar columna de profundidad, en magnitud de profundidad y unidad de medición**

Esta función separa la columna de profundidad en una de magnitud y la otra con la unidad de medición. Separa la columna 'PROFUNDIDAD' en 'PROFUNDIDAD_VALOR' (numérica) y 'PROFUNDIDAD_UNIDAD' (texto).

```python
#Descomponer columna de profundidad
def separar_profundidad(df):
    #Copia del df original
    df = df.copy()
    #Transformación a str de la columna
    df['PROFUNDIDAD'] = df['PROFUNDIDAD'].astype(str)
    #Extracción de valores númericos
    df['PROFUNDIDAD_VALOR'] = df['PROFUNDIDAD'].str.extract(r'(\d+\.?\d*)', expand=False)
    #transformación a númerico
    df['PROFUNDIDAD_VALOR'] = pd.to_numeric(df['PROFUNDIDAD_VALOR'], errors='coerce')
    #Extracción de valores de texto
    df['PROFUNDIDAD_UNIDAD'] = df['PROFUNDIDAD'].str.extract(r'\d+\.?\d*\s*(.*)', expand=False)
    #Quita caracteres vacios del texto
    df['PROFUNDIDAD_UNIDAD'] = df['PROFUNDIDAD_UNIDAD'].str.strip()
    print(df[['PROFUNDIDAD_VALOR', 'PROFUNDIDAD_UNIDAD']].head())
    return df
```

###**8.5) Separar columna de locación en distancia al epicentro, dirección y nombre del lugar desde donde se mide la distancia**

Esta función separa la columna ubicación en tres columnas: Distancia_epicentro_km, dirección y nombre_lugar.

```python
def separar_ubicacion(df):
    #Copiar df original
    df = df.copy()
    #Iniciar nuevas columnas
    df['DISTANCIA_EPICENTRO_KM'] = pd.NA
    df['DIRECCION'] = pd.NA
    df['NOMBRE_LUGAR'] = pd.NA
    #Transformar valor a string
    ubicacion_str = df['UBICACION'].astype(str)
    #Patrón de extracción
    patron = r'(\d+)\s*km\s*al\s*([A-Za-z\s]+?)\s*de\s*(.+)'
    #Aplica patrón de busqueda
    busqueda_patron = ubicacion_str.str.extract(patron, expand=True)
    #Identifica las filas correspondientes al patrón
    mask_full = busqueda_patron[0].notna()
    #Asigna las columnas para guardar los resultados de la búsqueda
    df.loc[mask_full, 'DISTANCIA_EPICENTRO_KM'] = pd.to_numeric(busqueda_patron.loc[mask_full, 0], errors='coerce')
    df.loc[mask_full, 'DIRECCION'] = busqueda_patron.loc[mask_full, 1].str.strip()
    df.loc[mask_full, 'NOMBRE_LUGAR'] = busqueda_patron.loc[mask_full, 2].str.strip()
    #Encuentra filas que no coinciden con el patrón
    mask_remaining = df['DISTANCIA_EPICENTRO_KM'].isna() & df['NOMBRE_LUGAR'].isna()
    dist_loc_matches = ubicacion_str[mask_remaining].str.extract(r'(\d+)\s*km\s*de\s*(.+)', expand=True)
    mask_dist_loc = dist_loc_matches[0].notna()
    #Asigna las columnas para guardar los resultados de la búsqueda
    df.loc[mask_remaining & mask_dist_loc, 'DISTANCIA_EPICENTRO_KM'] = pd.to_numeric(dist_loc_matches.loc[mask_dist_loc, 0], errors='coerce')
    df.loc[mask_remaining & mask_dist_loc, 'DIRECCION'] = 'Directo'
    df.loc[mask_remaining & mask_dist_loc, 'NOMBRE_LUGAR'] = dist_loc_matches.loc[mask_dist_loc, 1].str.strip()
    #Filas sin ninguna coincidencia previa se guardan como texto libre
    mask_final_remaining = df['DISTANCIA_EPICENTRO_KM'].isna() & df['NOMBRE_LUGAR'].isna()
    df.loc[mask_final_remaining, 'NOMBRE_LUGAR'] = ubicacion_str[mask_final_remaining].str.strip()

    return df
```

###**8.6) Separar columna de magnitud, en magnitud y unidad de medición.**

Función para separar columna de profundidad en magnitud y unidad de medición.



```python
#Descomponer columna de magnitud
def separar_magnitud(df):
    #Copia del df original
    df = df.copy()
    #Transformación de valores de la columna a string
    df['MAGNITUD'] = df['MAGNITUD'].astype(str)
    #Asignar columna magnitud_valor con patrón de búsqueda
    df['MAGNITUD_VALOR'] = df['MAGNITUD'].str.extract(r'(\d+\.?\d*)', expand=False)
    #Transformación a valor númerico
    df['MAGNITUD_VALOR'] = pd.to_numeric(df['MAGNITUD_VALOR'], errors='coerce')
    #Asignar columna magnitud_unidad con patrón de búsqueda
    df['MAGNITUD_UNIDAD'] = df['MAGNITUD'].str.extract(r'\d+\.?\d*\s*(.*)', expand=False)
    #Limpieza de espacios vacios de la columna magnitud_unidad
    df['MAGNITUD_UNIDAD'] = df['MAGNITUD_UNIDAD'].str.strip()
    return df
```

###**8.7) transformación de valores de las diferentes escalas sismicas.**

Las magnitudes presentan las siguiente nomenclaturas de escala:

1) Ml
2) Mw
3) Mww
4) Mb
5) Mlv
6) M
7) Ms
8) Mc

Mww es una variante de Mw, se trató como Mw.M equivale a que no se registro el tipo de escala por lo tanto, es Mw. Mlv es el componente vertical de la escala Ml, por lo tanto se tratará como Ml. En este caso se debe hacer la conversión a Mw (magnitud de momento) de las diferentes escalas.

Fórmulas de equivalencia:

1) Mw=0.85Ml+1.03(Scordilis, 2006)

2) Mw=0.85Mb+1.03(Scordilis, 2006)

3) Mw=0.85Mc+1.03(Scordilis, 2006)

4) Mw= 0.67Ms+2.07(Scordilis, 2006)


Scordilis, E. M. (2006). Empirical global relations converting MS​ and mb​ to moment magnitude. Journal of Seismology, 10(2), 225–236. https://doi.org/10.1007/s10950-006-9012-







Función para estandarizar escala de sismicidad

```python
#homogenización de la escala sísmica
def estandarizar_magnitud_mw(df):
    #crear copia del df original
    df = df.copy()
    # Inicializar la nueva columna MAGNITUD_MW con los valores originales.
    # Las magnitudes ya en Mw, M, Mww mantendrán su valor original.
    df['MAGNITUD_MW'] = df['MAGNITUD_VALOR']
     # Asegura tipo string y recategoriza 'Mlv' como 'Ml'
    df['MAGNITUD_UNIDAD'] = df['MAGNITUD_UNIDAD'].astype(str)
    mlv = df['MAGNITUD_UNIDAD'] == 'Mlv'
    df.loc[mlv, 'MAGNITUD_UNIDAD'] = 'Ml'
    print(f"Se han recategorizado {mlv.sum()} magnitudes de 'Mlv' a 'Ml'.")
    # Conversión para Ml (y Mlv)
    ml = df['MAGNITUD_UNIDAD'] == 'Ml'
    df.loc[ml, 'MAGNITUD_MW'] = 0.85 * df.loc[ml, 'MAGNITUD_VALOR'] + 1.03
    print(f"Se han convertido {ml.sum()} magnitudes de 'Ml' a 'Mw'.")
    # Conversión para Mb
    mb = df['MAGNITUD_UNIDAD'] == 'Mb'
    df.loc[mb, 'MAGNITUD_MW'] = 0.85 * df.loc[mb, 'MAGNITUD_VALOR'] + 1.03
    print(f"Se han convertido {mb.sum()} magnitudes de 'Mb' a 'Mw'.")
    # Conversión para Mc
    mc = df['MAGNITUD_UNIDAD'] == 'Mc'
    df.loc[mc, 'MAGNITUD_MW'] = 0.85 * df.loc[mc, 'MAGNITUD_VALOR'] + 1.03
    print(f"✓ Se han convertido {mc.sum()} magnitudes de 'Mc' a 'Mw'.")
    # Conversión para Ms
    ms = df['MAGNITUD_UNIDAD'] == 'Ms'
    df.loc[ms, 'MAGNITUD_MW'] = 0.67 * df.loc[ms, 'MAGNITUD_VALOR'] + 2.07
    print(f"✓ Se han convertido {ms.sum()} magnitudes de 'Ms' a 'Mw'.")
    print("\nPrimeras filas con la nueva columna MAGNITUD_MW:")
    print(df[['MAGNITUD_VALOR', 'MAGNITUD_UNIDAD', 'MAGNITUD_MW']].head())
    return df
```

###**8.8) Eliminar registros sismicos que no hayan ocurrido en Chile**

Esta función detecta los sismos que están dentro de un área determinada. Los que esta fuera del área de trabajo, son eliminados.

```python
def detectar_eventos_fuera_de_chile(df, lat_min=-56.0, lat_max=-17.0, lon_min=-90.0, lon_max=-66.0, n_ejemplos=10):
    """
    Detecta registros sísmicos fuera del rango geográfico de Chile y devuelve
    un DataFrame con los eventos que SÍ ocurrieron dentro de Chile.

    Parámetros
    ----------
    df : pd.DataFrame
        DataFrame con columnas 'LATITUD', 'LONGITUD', etc.
    lat_min : float
        Latitud mínima del bounding box (default: -56.0).
    lat_max : float
        Latitud máxima del bounding box (default: -17.0).
    lon_min : float
        Longitud mínima del bounding box (default: -90.0).
    lon_max : float
        Longitud máxima del bounding box (default: -66.0).
    n_ejemplos : int
        (No usado en la versión actual, ya que se eliminan los fuera de Chile).

    Retorna
    -------
    pd.DataFrame
        Subconjunto de df con los registros DENTRO del rango geográfico de Chile.
    """
    # Máscara para identificar eventos FUERA del rango chileno
    mascara_fuera = (
        (df['LATITUD'] < lat_min) | (df['LATITUD'] > lat_max) |
        (df['LONGITUD'] < lon_min) | (df['LONGITUD'] > lon_max)
    )
    # Copiar dataframe
    df_copia = df.copy()
    # DataFrame con eventos DENTRO del rango chileno
    df_en_chile = df_copia[~mascara_fuera]
    n_total = len(df_copia)
    n_fuera = mascara_fuera.sum() # Suma los True de la máscara
    n_en_chile = len(df_en_chile)
    porcentaje_fuera = (n_fuera / n_total) * 100
    print(f"Total de registros iniciales: {n_total}")
    print(f"Registros fuera del rango chileno (eliminados): {n_fuera} ({porcentaje_fuera:.2f}%)")
    print(f"Registros dentro del rango chileno (conservados): {n_en_chile}\n")
    # Eliminar la columna 'UBICACION' si existe y no se necesita, ahora aplicada al df filtrado
    df_en_chile = df_en_chile.drop(columns=['UBICACION'], errors='ignore')
    return df_en_chile
```

```python
#Ejecutar pipeline
def ejecutar_pipeline(ruta_archivo):

    print("--- Iniciando pipeline de procesamiento---")
    # Carga de datos inicial
    df = pd.read_csv(ruta_archivo)
    print(f'\n✓ Dataset cargado: {df.shape[0]:,} registros × {df.shape[1]} columnas')
    explorar_datos(df) # Llamar a la función de exploración para el estado inicial
    # 8.1) Renombrar variables del dataset a español
    df = renombrar_variables(df)
    print("\n--- 8.1 Renombrado de variables completado ---")
    # 8.2) Limpieza de datos (eliminar duplicados)
    df = limpiar_duplicados(df)
    print("--- 8.2 Limpieza de duplicados completado ---")
    # 8.3) Conversión de columnas de fecha a formato datetime de variable Fecha UTC
    df = procesar_fecha_utc(df)
    print("--- 8.3 Procesamiento de fecha UTC completado ---")
    # 8.4) Separar columna de profundidad
    df = separar_profundidad(df)
    print("--- 8.4 Separación de profundidad completado ---")
    # 8.5) Separar columna de locación (prerequisito para filtrado geográfico)
    df = separar_ubicacion(df)
    print("--- 8.5 Separación de ubicación completado ---")
    # 8.6) Separar columna de magnitud, en magnitud y unidad de medición.
    df = separar_magnitud(df)
    print("--- 8.6 Separar columna de magnitud, en magnitud y unidad de medición---")
    # 8.7) transformación de valores de las diferentes escalas sismicas.
    df = estandarizar_magnitud_mw(df)
    print("--- 8.7 transformación de valores de las diferentes escalas sismicas. ---")
    # 8.8) Eliminar registros sísmicos que no hayan ocurrido en Chile (Geographical Filtering)
    df = detectar_eventos_fuera_de_chile(df)
    print("--- 8.8) Eliminar registros sísmicos que no hayan ocurrido en Chile (Geographical Filtering) ---")
    print("--- Pipeline de procesamiento hasta la Opción 8.8 completado ---")
    return df
# Ejecutar pipeline
df_procesado = ejecutar_pipeline(ruta)

print("\nDataFrame resultante después del pipeline (primeras 5 filas):")
print(df_procesado.head())
print(f"\nDimensiones finales del DataFrame: {df_procesado.shape[0]:,} registros × {df_procesado.shape[1]} columnas")
```

###**8.13)categorizar profundidad y magnitud_MW**

```python
def categorizar_profundidad(df):
    # Definir los límites y etiquetas para la categorización de PROFUNDIDAD_VALOR
    # Basado en escalas geológicas estándar:
    # Somero: 0-70 km
    # Intermedio: 70-300 km
    # Profundo: >300 km
    bins = [0, 70, 300, np.inf]
    labels = ['somero', 'intermedio', 'profundo']
    df['CATEGORIA_PROFUNDIDAD'] = pd.cut(df['PROFUNDIDAD_VALOR'], bins=bins, labels=labels, right=False, include_lowest=True)
    print("--- Categorización de PROFUNDIDAD_VALOR completada ---")
    print(df[['PROFUNDIDAD_VALOR', 'CATEGORIA_PROFUNDIDAD']].head())
    print("Conteo de categorías de profundidad:\n", df['CATEGORIA_PROFUNDIDAD'].value_counts())
    return df

def categorizar_magnitud(df):
    # Definir los límites y etiquetas para la categorización de MAGNITUD_MW
    # Basado en la escala de Richter (aproximación para Mw):
    # Micro: < 2.0
    # Menor: 2.0 - 3.9
    # Ligero: 4.0 - 4.9
    # Moderado: 5.0 - 5.9
    # Fuerte: 6.0 - 6.9
    # Mayor: 7.0 - 7.9
    # Grande: >= 8.0
    bins = [0, 2.0, 4.0, 5.0, 6.0, 7.0, 8.0, np.inf]
    labels = ['Micro', 'Menor', 'Ligero', 'Moderado', 'Fuerte', 'Mayor', 'Grande']
    df['CATEGORIA_MAGNITUD'] = pd.cut(df['MAGNITUD_MW'], bins=bins, labels=labels, right=False, include_lowest=True)
    print("\n--- Categorización de MAGNITUD_MW completada ---")
    print(df[['MAGNITUD_MW', 'CATEGORIA_MAGNITUD']].head())
    print("Conteo de categorías de magnitud:\n", df['CATEGORIA_MAGNITUD'].value_counts())
    return df

def categorizar_distancia_epicentro(df):
    # Asegurarse de que la columna DISTANCIA_EPICENTRO_KM sea numérica
    # Si es 'object' (como se vio en info()), se intenta convertir a numérica
    if pd.api.types.is_object_dtype(df['DISTANCIA_EPICENTRO_KM']):
        df['DISTANCIA_EPICENTRO_KM'] = pd.to_numeric(df['DISTANCIA_EPICENTRO_KM'], errors='coerce')
        # Rellenar NaNs resultantes de la conversión con la media o mediana si es necesario, o descartar filas
        # Por simplicidad, aquí asumimos que los NaNs se manejarán más tarde o ya fueron manejados.

    # Definir los límites y etiquetas para la categorización de DISTANCIA_EPICENTRO_KM
    # Usando rangos comunes y similares a profundidad para consistencia:
    # Cercano: 0-70 km
    # Intermedio: 70-300 km
    # Lejano: >300 km
    bins = [0, 70, 300, np.inf]
    labels = ['Cercano', 'Intermedio', 'Lejano']
    df['CATEGORIA_DISTANCIA_EPICENTRO'] = pd.cut(df['DISTANCIA_EPICENTRO_KM'], bins=bins, labels=labels, right=False, include_lowest=True)
    print("\n--- Categorización de DISTANCIA_EPICENTRO_KM completada ---")
    print(df[['DISTANCIA_EPICENTRO_KM', 'CATEGORIA_DISTANCIA_EPICENTRO']].head())
    print("Conteo de categorías de distancia:\n", df['CATEGORIA_DISTANCIA_EPICENTRO'].value_counts())
    return df

# Aplicar las funciones al DataFrame df_final
# Asegúrate de que df_final esté disponible en el entorno
if 'df_final' in globals():
    df_final = categorizar_profundidad(df_final)
    df_final = categorizar_magnitud(df_final)
    df_final = categorizar_distancia_epicentro(df_final)
    print("\n--- df_final con nuevas columnas de categorización ---")
    print(df_final.head())
    print("\nInfo del DataFrame final con las nuevas categorías:")
    df_final.info()
else:
    print("Advertencia: 'df_final' no está definido.")
    print("Por favor, asegúrate de ejecutar las celdas anteriores que crean 'df_final'.")
```

###**8.14) Se aplica OneHotEncoder para el df limpio**

```python
def aplicar_one_hot_encoding(df):
    print("--- Iniciando One-Hot Encoding ---")
    # Identificar columnas categóricas (tipo 'object' o 'category')
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns

    if len(categorical_cols) == 0:
        print("No se encontraron columnas categóricas para aplicar One-Hot Encoding.")
        return df

    print(f"Columnas categóricas identificadas: {list(categorical_cols)}")

    # Aplicar One-Hot Encoding
    df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

    print(f"DataFrame original: {df.shape[1]} columnas")
    print(f"DataFrame después de One-Hot Encoding: {df_encoded.shape[1]} columnas")
    print("--- One-Hot Encoding finalizado ---")
    return df_encoded



if 'df_final' in globals():
    df_final_encoded = aplicar_one_hot_encoding(df_final)
    print("\nPrimeras 5 filas del DataFrame con One-Hot Encoding:")
    print(df_final_encoded.head())
    print("\nInformación del DataFrame codificado:")
    df_final_encoded.info()
```
