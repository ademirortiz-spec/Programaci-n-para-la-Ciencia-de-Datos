````python
from graphviz import Digraph
import os

# Crear directorio para guardar gráficos
if not os.path.exists('graficos'):
    os.makedirs('graficos')

# Crear el diagrama del mapa conceptual
dot = Digraph(comment='Mapa Conceptual - Dataset Terremotos', format='png')
dot.attr(rankdir='TB')
dot.attr('node', shape='box', style='rounded,filled', fillcolor='lightblue', fontname='Arial')
dot.attr('graph', bgcolor='white', fontname='Arial')

# Nodo raíz
dot.node('root', 'DATASET TERREMOTOS\nCHILE 2000-2024\n(134,062 registros)', 
         shape='box', style='rounded,filled', fillcolor='#FFD700', fontsize='12', fontweight='bold')

# Variables Categóricas
dot.node('cat', 'VARIABLES\nCATEGORICAS\n(6 variables)', 
         fillcolor='#87CEEB', fontsize='11')
dot.node('nom', 'NOMINALES\n(Sin orden)', 
         fillcolor='#B0E0E6', fontsize='10')
dot.node('ord', 'ORDINALES\n(Con orden)', 
         fillcolor='#B0E0E6', fontsize='10')

# Variables Numéricas
dot.node('num', 'VARIABLES\nNUMERICAS\n(2 variables)', 
         fillcolor='#90EE90', fontsize='11')
dot.node('cont', 'CONTINUAS\n(Infinitos valores)', 
         fillcolor='#98FB98', fontsize='10')

# Nominales - Variables específicas
dot.node('nom1', 'Location\n(28,965 valores únicos)', fillcolor='#E0FFFF', fontsize='9')
dot.node('nom2', 'UTC Date\n(133,946 valores únicos)', fillcolor='#E0FFFF', fontsize='9')

# Ordinales - Variables específicas
dot.node('ord1', 'Magnitude\n(244 valores únicos)', fillcolor='#E0FFFF', fontsize='9')
dot.node('ord2', 'Profoundity\n(391 valores únicos)', fillcolor='#E0FFFF', fontsize='9')
dot.node('ord3', 'Date\n(8,771 valores únicos)', fillcolor='#E0FFFF', fontsize='9')
dot.node('ord4', 'Hour\n(68,078 valores únicos)', fillcolor='#E0FFFF', fontsize='9')

# Continuas - Variables específicas
dot.node('cont1', 'Latitude\n(Rango: -65.4 a 58.7)', fillcolor='#F0FFF0', fontsize='9')
dot.node('cont2', 'Longitude\n(Rango: -179.9 a 179.6)', fillcolor='#F0FFF0', fontsize='9')

# Conexiones principales
dot.edge('root', 'cat')
dot.edge('root', 'num')

# Conexiones categóricas
dot.edge('cat', 'nom')
dot.edge('cat', 'ord')

# Conexiones nominales
dot.edge('nom', 'nom1')
dot.edge('nom', 'nom2')

# Conexiones ordinales
dot.edge('ord', 'ord1')
dot.edge('ord', 'ord2')
dot.edge('ord', 'ord3')
dot.edge('ord', 'ord4')

# Conexiones numéricas
dot.edge('num', 'cont')
dot.edge('cont', 'cont1')
dot.edge('cont', 'cont2')

# Guardar el gráfico
output_path = dot.render('graficos/mapa_conceptual_option2', cleanup=True)
print(f"✓ Mapa conceptual guardado en: {output_path}")
print(f"✓ Archivo PNG creado: graficos/mapa_conceptual_option2.png")
````

# Análisis de Terremotos en Chile (2000-2024)

Este notebook contiene un análisis exploratorio de datos sísmicos en Chile entre los años 2000 y 2024.

**Fuente de datos:** [Kaggle - Earthquakes in Chile 2000-2024](https://www.kaggle.com/datasets/javierquinterosm/earthquakes-in-chile-2000-2024)

---

## 1. Importación de Librerías

Importamos las librerías necesarias para el análisis de datos.

````python
# Librerías para manipulación de datos
import pandas as pd
import numpy as np

# Librerías para visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de visualización
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
%matplotlib inline

# Configuración de pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("✓ Librerías importadas correctamente")
````

## 2. Carga de Datos

Cargamos el dataset de terremotos desde el archivo CSV.

````python
# Cargar datos
df = pd.read_csv('dataset/EarthquakesChile_2000-2024.csv')

# Mostrar información básica
print(f"Dimensiones del dataset: {df.shape}")
print(f"\nPrimeras filas:")
df.head()
````

## 3. Exploración Inicial de Datos

Analizamos la estructura y características básicas del dataset.

````python
# Información del dataset
print("Información del Dataset:")
print(df.info())

print("\n" + "="*50)
print("\nEstadísticas Descriptivas:")
df.describe()
````

````python
# Verificar valores nulos
print("Valores Nulos por Columna:")
null_counts = df.isnull().sum()
null_percentages = (df.isnull().sum() / len(df)) * 100

null_df = pd.DataFrame({
    'Cantidad': null_counts,
    'Porcentaje': null_percentages
})

null_df[null_df['Cantidad'] > 0]
````

## 4. Limpieza y Transformación de Datos

Preparamos los datos para el análisis.

````python
# Convertir columnas de fecha a datetime
df['UTC Date'] = pd.to_datetime(df['UTC Date'])
df['Date'] = pd.to_datetime(df['Date'])

# Extraer componentes temporales
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek

print("✓ Transformaciones aplicadas correctamente")
df.head()
````

## 5. Análisis Exploratorio

### 5.1 Distribución de Magnitudes

````python
# Histograma de magnitudes
plt.figure(figsize=(12, 6))
plt.hist(df['Magnitude'].dropna(), bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('Magnitud')
plt.ylabel('Frecuencia')
plt.title('Distribución de Magnitudes de Terremotos en Chile (2000-2024)')
plt.grid(True, alpha=0.3)
plt.show()

# Estadísticas de magnitud
print(f"Magnitud promedio: {df['Magnitude'].mean():.2f}")
print(f"Magnitud máxima: {df['Magnitude'].max():.2f}")
print(f"Magnitud mínima: {df['Magnitude'].min():.2f}")
````

### 5.2 Terremotos por Año

````python
# Contar terremotos por año
earthquakes_per_year = df.groupby('Year').size()

plt.figure(figsize=(14, 6))
earthquakes_per_year.plot(kind='bar', color='steelblue', edgecolor='black')
plt.xlabel('Año')
plt.ylabel('Número de Terremotos')
plt.title('Número de Terremotos por Año en Chile (2000-2024)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
````

### 5.3 Distribución de Profundidad

````python
# Boxplot de profundidad
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, y='Profoundity', color='coral')
plt.ylabel('Profundidad (km)')
plt.title('Distribución de Profundidad de Terremotos')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
````

### 5.4 Mapa de Distribución Geográfica

````python
# Scatter plot de ubicaciones
plt.figure(figsize=(10, 14))
scatter = plt.scatter(df['Longitude'], df['Latitude'], 
                     c=df['Magnitude'], cmap='YlOrRd', 
                     s=10, alpha=0.5, edgecolors='none')
plt.colorbar(scatter, label='Magnitud')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Distribución Geográfica de Terremotos en Chile (2000-2024)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

## 6. Análisis de Correlaciones

Analizamos las relaciones entre variables numéricas.

````python
# Matriz de correlación
numeric_cols = ['Profoundity', 'Magnitude', 'Latitude', 'Longitude']
correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', 
            center=0, square=True, linewidths=1)
plt.title('Matriz de Correlación - Variables Sísmicas')
plt.tight_layout()
plt.show()
````
