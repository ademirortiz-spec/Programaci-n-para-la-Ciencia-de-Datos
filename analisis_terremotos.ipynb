# Análisis Exploratorio de Terremotos en Chile (2000-2024)

## 📋 Introducción

Chile es uno de los países con mayor actividad sísmica en el mundo, ubicado en la zona de subducción de la placa de Nazca y la placa Sudamericana. Este notebook presenta un análisis exploratorio de 134,062 eventos sísmicos registrados entre 2000 y 2024.

**Objetivo:** Identificar patrones, tendencias y características de la actividad sísmica en territorio chileno.

**Fuente de datos:** [Kaggle - Earthquakes in Chile 2000-2024](https://www.kaggle.com/datasets/javierquinterosm/earthquakes-in-chile-2000-2024)

**Variables disponibles:** 8 (6 categóricas, 2 numéricas)

## 1. Importación de Librerías

Configuramos el entorno de trabajo importando las librerías necesarias para análisis de datos y visualización.

````python
# Librerías para manipulación de datos
import pandas as pd
import numpy as np

# Librerías para visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de reproducibilidad
np.random.seed(42)  # Seed fijo para resultados reproducibles

# Configuración de visualización
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
%matplotlib inline

# Configuración de pandas para mejor visualización
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', None)

print("✓ Entorno configurado correctamente")
print(f"✓ Versiones:")
print(f"  - pandas: {pd.__version__}")
print(f"  - numpy: {np.__version__}")
print(f"  - matplotlib: {plt.matplotlib.__version__}")
print(f"  - seaborn: {sns.__version__}")
print(f"✓ Seed fijo en 42 para reproducibilidad")
````

## 2. Carga de Datos

Importamos el dataset desde el archivo CSV. Este archivo contiene 134,062 registros de eventos sísmicos.

**Ubicación del archivo:** `dataset/EarthquakesChile_2000-2024.csv`

**Estructura esperada:**
- Registros: 134,062 filas (eventos sísmicos)
- Variables: 8 columnas (6 categóricas, 2 numéricas)

````python
# Cargar datos desde archivo CSV (ruta relativa)
df = pd.read_csv('dataset/EarthquakesChile_2000-2024.csv')

# Mostrar información básica
print(f"{'='*60}")
print(f"INFORMACIÓN GENERAL DEL DATASET")
print(f"{'='*60}")
print(f"\n📊 Dimensiones: {df.shape[0]:,} registros × {df.shape[1]} variables")
print(f"\n📋 Primeras 5 filas del dataset:\n")
print(df.head())
````

## 3. Exploración Inicial de Datos

Analizamos la estructura del dataset para entender tipos de datos, valores faltantes y características principales.

````python
# Información detallada del dataset
print(f"\n{'='*60}")
print("TIPOS DE DATOS Y ESTRUCTURA")
print(f"{'='*60}\n")
print(df.info())

print(f"\n{'='*60}")
print("ESTADÍSTICAS DESCRIPTIVAS")
print(f"{'='*60}\n")
print(df.describe())
````

````python
# Análisis de valores faltantes
print(f"\n{'='*60}")
print("VALORES FALTANTES")
print(f"{'='*60}\n")

null_counts = df.isnull().sum()
null_percentages = (df.isnull().sum() / len(df)) * 100

null_df = pd.DataFrame({
    'Columna': null_counts.index,
    'Cantidad_Nulos': null_counts.values,
    'Porcentaje': null_percentages.values
})

null_df_missing = null_df[null_df['Cantidad_Nulos'] > 0]

if len(null_df_missing) > 0:
    print(null_df_missing.to_string(index=False))
else:
    print("✓ No hay valores faltantes en el dataset")
````

## 4. Limpieza y Transformación de Datos

Preparamos los datos para análisis posteriores. Las transformaciones incluyen:
- Conversión de columnas de fecha a formato datetime
- Extracción de componentes temporales (año, mes, día, día de la semana)
- Esto permitirá realizar análisis temporales más profundos

````python
# Convertir columnas de fecha a datetime (formato estándar)
df['UTC Date'] = pd.to_datetime(df['UTC Date'])
df['Date'] = pd.to_datetime(df['Date'])

# Extraer componentes temporales para análisis
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek

print(f"{'='*60}")
print("TRANSFORMACIÓN DE DATOS")
print(f"{'='*60}")
print(f"\n✓ Conversión de fechas completada")
print(f"✓ Componentes temporales extraídos: Year, Month, Day, DayOfWeek")
print(f"\nPrimeras filas después de transformación:\n")
print(df.head())
````

## 5. Análisis Exploratorio

### 5.1 Distribución de Magnitudes

La magnitud es una medida de la energía liberada por un terremoto. En esta sección analizamos cómo se distribuyen los eventos según su magnitud.

````python
# Histograma de magnitudes con análisis descriptivo
plt.figure(figsize=(12, 6))
plt.hist(df['Magnitude'].dropna(), bins=50, edgecolor='black', alpha=0.7, color='steelblue')
plt.xlabel('Magnitud (Escala Richter)', fontsize=11, fontweight='bold')
plt.ylabel('Frecuencia (número de eventos)', fontsize=11, fontweight='bold')
plt.title('Distribución de Magnitudes de Terremotos en Chile (2000-2024)', fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Estadísticas de magnitud con contexto
print(f"\n{'='*60}")
print("ANÁLISIS DE MAGNITUDES")
print(f"{'='*60}\n")

mag_stats = df['Magnitude'].dropna()
print(f"Total de eventos con magnitud registrada: {len(mag_stats):,}")
print(f"Magnitud promedio: {mag_stats.mean():.2f}")
print(f"Magnitud mediana: {mag_stats.median():.2f}")
print(f"Magnitud máxima: {mag_stats.max():.2f}")
print(f"Magnitud mínima: {mag_stats.min():.2f}")
print(f"Desviación estándar: {mag_stats.std():.2f}")

# Distribución por rangos de magnitud
print(f"\n📊 Distribución por rangos de magnitud:")
mag_ranges = [(0, 3), (3, 4), (4, 5), (5, 6), (6, 7), (7, 9)]
for min_mag, max_mag in mag_ranges:
    count = len(df[(df['Magnitude'] >= min_mag) & (df['Magnitude'] < max_mag)])
    percentage = (count / len(mag_stats)) * 100
    print(f"  {min_mag}.0 - {max_mag}.0: {count:>6,} eventos ({percentage:>5.1f}%)")
````

### 5.2 Terremotos por Año

Analizamos la tendencia temporal: ¿ha aumentado o disminuido la cantidad de eventos sísmicos registrados a lo largo de los 24 años?

````python
# Contar terremotos por año
earthquakes_per_year = df.groupby('Year').size()

plt.figure(figsize=(14, 6))
earthquakes_per_year.plot(kind='bar', color='steelblue', edgecolor='black', linewidth=1.5)
plt.xlabel('Año', fontsize=11, fontweight='bold')
plt.ylabel('Número de Terremotos', fontsize=11, fontweight='bold')
plt.title('Número de Terremotos Registrados por Año en Chile (2000-2024)', fontsize=13, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Análisis de tendencia
print(f"\n{'='*60}")
print("ANÁLISIS TEMPORAL - TERREMOTOS POR AÑO")
print(f"{'='*60}\n")
print(f"Año con mayor actividad: {earthquakes_per_year.idxmax()} ({earthquakes_per_year.max():,} eventos)")
print(f"Año con menor actividad: {earthquakes_per_year.idxmin()} ({earthquakes_per_year.min():,} eventos)")
print(f"Promedio de eventos por año: {earthquakes_per_year.mean():,.0f}")
print(f"\nPromedio (2000-2012): {earthquakes_per_year[2000:2012].mean():,.0f} eventos/año")
print(f"Promedio (2013-2024): {earthquakes_per_year[2013:2024].mean():,.0f} eventos/año")
````

### 5.3 Distribución de Profundidad

La profundidad de un terremoto (en km) es importante para entender la actividad de diferentes capas terrestres. Analizamos cómo se distribuyen los eventos a diferentes profundidades.

````python
# Boxplot de profundidad con análisis
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, y='Profoundity', color='coral')
plt.ylabel('Profundidad (km)', fontsize=11, fontweight='bold')
plt.title('Distribución de Profundidad de Terremotos en Chile', fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Estadísticas de profundidad
print(f"\n{'='*60}")
print("ANÁLISIS DE PROFUNDIDAD")
print(f"{'='*60}\n")

prof_stats = df['Profoundity'].dropna()
print(f"Total de eventos con profundidad registrada: {len(prof_stats):,}")
print(f"Profundidad promedio: {prof_stats.mean():.2f} km")
print(f"Profundidad mediana: {prof_stats.median():.2f} km")
print(f"Profundidad máxima: {prof_stats.max():.2f} km")
print(f"Profundidad mínima: {prof_stats.min():.2f} km")
print(f"Rango intercuartil (Q3-Q1): {prof_stats.quantile(0.75) - prof_stats.quantile(0.25):.2f} km")

# Clasificación por profundidad
print(f"\n📊 Clasificación de eventos por profundidad:")
prof_ranges = [
    (0, 70, "Superficiales"),
    (70, 300, "Intermedios"),
    (300, 700, "Profundos")
]
for min_prof, max_prof, tipo in prof_ranges:
    count = len(df[(df['Profoundity'] >= min_prof) & (df['Profoundity'] < max_prof)])
    percentage = (count / len(prof_stats)) * 100
    print(f"  {tipo:15} ({min_prof}-{max_prof} km): {count:>6,} eventos ({percentage:>5.1f}%)")
````

### 5.4 Mapa de Distribución Geográfica

Visualizamos la ubicación de todos los terremotos en el territorio chileno, coloreados según su magnitud. Esto permite identificar zonas de mayor concentración de actividad sísmica.

````python
# Scatter plot de ubicaciones geográficas
plt.figure(figsize=(10, 14))
scatter = plt.scatter(df['Longitude'], df['Latitude'], 
                     c=df['Magnitude'], cmap='YlOrRd', 
                     s=10, alpha=0.5, edgecolors='none')
colorbar = plt.colorbar(scatter, label='Magnitud')
plt.xlabel('Longitud', fontsize=11, fontweight='bold')
plt.ylabel('Latitud', fontsize=11, fontweight='bold')
plt.title('Distribución Geográfica de Terremotos en Chile (2000-2024)\nColores indican magnitud', 
          fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Análisis geográfico
print(f"\n{'='*60}")
print("ANÁLISIS GEOGRÁFICO")
print(f"{'='*60}\n")

lat_stats = df['Latitude'].dropna()
lon_stats = df['Longitude'].dropna()

print(f"Rango de Latitud: {lat_stats.min():.2f}° a {lat_stats.max():.2f}°")
print(f"Rango de Longitud: {lon_stats.min():.2f}° a {lon_stats.max():.2f}°")
print(f"\nCentro aproximado del área analizada:")
print(f"  Latitud: {lat_stats.mean():.2f}°")
print(f"  Longitud: {lon_stats.mean():.2f}°")
````

## 6. Análisis de Correlaciones

Investigamos las relaciones estadísticas entre variables numéricas. Una matriz de correlación nos ayuda a identificar qué variables se mueven juntas.

````python
# Matriz de correlación entre variables numéricas
numeric_cols = ['Profoundity', 'Magnitude', 'Latitude', 'Longitude']
correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', 
            center=0, square=True, linewidths=1, fmt='.3f',
            cbar_kws={'label': 'Coeficiente de Correlación'})
plt.title('Matriz de Correlación - Variables Sísmicas', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# Interpretación de correlaciones
print(f"\n{'='*60}")
print("ANÁLISIS DE CORRELACIONES")
print(f"{'='*60}\n")

print("Pares de variables con correlación notable:\n")
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        var1 = correlation_matrix.columns[i]
        var2 = correlation_matrix.columns[j]
        corr_value = correlation_matrix.iloc[i, j]
        
        if abs(corr_value) > 0.3:  # Umbral de relevancia
            strength = "fuerte" if abs(corr_value) > 0.7 else "moderada"
            direction = "positiva" if corr_value > 0 else "negativa"
            print(f"  • {var1} vs {var2}: {corr_value:.3f} ({direction}, {strength})")
````
