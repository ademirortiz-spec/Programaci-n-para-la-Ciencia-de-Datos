##**5. Dataset**

El dataset cuenta con 8 variables en total. Cada variable representa lo siguiente:

1) UTC Date: Fecha y hora del evento sísmico, expresada en el Tiempo Universal Coordinado (UTC). Es la referencia estándar utilizada globalmente en sismología.

2) Profoundity (Profundidad): Distancia vertical (medida en kilómetros) desde la superficie de la Tierra hasta el hipocentro o foco exacto donde se originó la ruptura del sismo.

3) Magnitude (Magnitud): Medida cuantitativa de la energía liberada en el foco del sismo. Incluye el valor numérico y el tipo de escala empleada.

4) Date (Fecha local): La fecha calendario en la que se registró el sismo ajustada al huso horario local de Chile.

5) Hour (Hora local): La hora exacta del registro del sismo ajustada al huso horario local de Chile.

6) Location (Ubicación): Descripción geográfica o referencia textual que indica a cuántos kilómetros y en qué dirección cardinal se localizó el epicentro respecto a una comuna, ciudad o localidad de referencia.

7) Latitude (Latitud): Coordenada geográfica que indica la posición del epicentro al norte o al sur del ecuador terrestre (expresada en grados decimales, con valores negativos para el hemisferio sur).

8) Longitude (Longitud): Coordenada geográfica que determina la posición del epicentro al este u oeste del meridiano de Greenwich (expresada en grados decimales, con valores negativos para el hemisferio occidental).

###**8.3) Conversión de columnas de fecha a formato datetime de variable Fecha UTC**



Esta función se genera una columna independiente para las variables año, mes y día de la ocurrencia de terremotos para el formato UTC.

````python
def procesar_fecha_utc(df):
    df = df.copy()
    df['FECHA_UTC'] = pd.to_datetime(df['FECHA_UTC'], errors='coerce')
    df['ANIO_UTC'] = df['FECHA_UTC'].dt.year
    df['MES_UTC'] = df['FECHA_UTC'].dt.month
    df['DIA_UTC'] = df['FECHA_UTC'].dt.day
    df['HORA_UTC'] = df['FECHA_UTC'].dt.strftime('%H:%M:%S')

    # Displaying relevant columns that this function modifies
    print(df[['FECHA_UTC', 'ANIO_UTC', 'MES_UTC', 'DIA_UTC', 'HORA_UTC']].head())
    return df
````

###**8.4) Separar columna de profundidad, en magnitud de profundidad y unidad de medición**

Esta función separa la columna de profundidad en una de magnitud y la otra con la unidad de medición. Separa la columna 'PROFUNDIDAD' en 'PROFUNDIDAD_VALOR' (numérica) y 'PROFUNDIDAD_UNIDAD' (texto).

````python
def separar_profundidad(df):
    df = df.copy()
    df['PROFUNDIDAD'] = df['PROFUNDIDAD'].astype(str)
    df['PROFUNDIDAD_VALOR'] = df['PROFUNDIDAD'].str.extract(r'(\d+\.?\d*)', expand=False)
    df['PROFUNDIDAD_VALOR'] = pd.to_numeric(df['PROFUNDIDAD_VALOR'], errors='coerce')
    df['PROFUNDIDAD_UNIDAD'] = df['PROFUNDIDAD'].str.extract(r'\d+\.?\d*\s*(.*)', expand=False)
    df['PROFUNDIDAD_UNIDAD'] = df['PROFUNDIDAD_UNIDAD'].str.strip()
    #df = df.drop(columns=['PROFUNDIDAD']) # This line is commented out as requested by the user
    print(df[['PROFUNDIDAD_VALOR', 'PROFUNDIDAD_UNIDAD']].head())
    return df
````

###**8.5) Separar columna de locación en distancia al epicentro, dirección y nombre del lugar desde donde se mide la distancia**

Esta función separa la columna ubicación en tres columnas:

Distancia_epicentro_km, dirección y nombre_lugar.

````python
import re

def separar_ubicacion(df):
    #Copiar df original
    df = df.copy()
    #Iniciar nuevas columnas
    df['DISTANCIA_EPICENTRO_KM'] = pd.NA
    df['DIRECCION'] = pd.NA
    df['NOMBRE_LUGAR'] = pd.NA
    #Transformar valor a string
    ubicacion_str = df['UBICACION'].astype(str)
    patron = r'(\d+)\s*km\s*al\s*([A-Za-z\s]+?)\s*de\s*(.+)'
    full_matches = ubicacion_str.str.extract(patron, expand=True)
    mask_full = full_matches[0].notna()
    df.loc[mask_full, 'DISTANCIA_EPICENTRO_KM'] = pd.to_numeric(full_matches.loc[mask_full, 0], errors='coerce')
    df.loc[mask_full, 'DIRECCION'] = full_matches.loc[mask_full, 1].str.strip()
    df.loc[mask_full, 'NOMBRE_LUGAR'] = full_matches.loc[mask_full, 2].str.strip()
    mask_remaining = df['DISTANCIA_EPICENTRO_KM'].isna() & df['NOMBRE_LUGAR'].isna()

    if mask_remaining.any():
        dist_loc_matches = ubicacion_str[mask_remaining].str.extract(r'(\d+)\s*km\s*de\s*(.+)', expand=True)
        mask_dist_loc = dist_loc_matches[0].notna()

        df.loc[mask_remaining & mask_dist_loc, 'DISTANCIA_EPICENTRO_KM'] = pd.to_numeric(dist_loc_matches.loc[mask_dist_loc, 0], errors='coerce')
        df.loc[mask_remaining & mask_dist_loc, 'DIRECCION'] = 'Directo'
        df.loc[mask_remaining & mask_dist_loc, 'NOMBRE_LUGAR'] = dist_loc_matches.loc[mask_dist_loc, 1].str.strip()
    mask_final_remaining = df['DISTANCIA_EPICENTRO_KM'].isna() & df['NOMBRE_LUGAR'].isna()

    if mask_final_remaining.any():
        df.loc[mask_final_remaining, 'NOMBRE_LUGAR'] = ubicacion_str[mask_final_remaining].str.strip()
    df = df.drop(columns=['UBICACION'])
    #print(df_copia[['DISTANCIA_EPICENTRO_KM', 'DIRECCION', 'NOMBRE_LUGAR']].head())
    return df
````

###**8.6) Separar columna de magnitud, en magnitud y unidad de medición.**

Función para separar columna de profundidad en magnitud y unidad de medición

Normalizar las dierentes unidad de medición que existen y solo usar magnitud de momento (Mw)

````python
def separar_magnitud(df):
    df = df.copy()
    df['MAGNITUD'] = df['MAGNITUD'].astype(str)
    df['MAGNITUD_VALOR'] = df['MAGNITUD'].str.extract(r'(\d+\.?\d*)', expand=False)
    df['MAGNITUD_VALOR'] = pd.to_numeric(df['MAGNITUD_VALOR'], errors='coerce')
    df['MAGNITUD_UNIDAD'] = df['MAGNITUD'].str.extract(r'\d+\.?\d*\s*(.*)', expand=False)
    df['MAGNITUD_UNIDAD'] = df['MAGNITUD_UNIDAD'].str.strip()
    df = df.drop(columns=['MAGNITUD'])
    print(df['MAGNITUD_UNIDAD'].unique())
    print(df[['MAGNITUD_VALOR', 'MAGNITUD_UNIDAD']].head())
    return df
````

###**8.7) transformación de valores de las diferentes escalas sismicas.**

Las magnitudes presentan las siguiente nomenclaturas de escala:

['Ml' 'Mw' 'Mww' nan 'Mb' 'Mlv' 'M' 'Ms' 'Mc']

Mw=0.85Ml+0.53

Mw=0.85Mb+1.03

Mw=0.85Mc+0.53

Mw= 0.67Ms+2.07

Mww = variante de Mw, sera tratada como Mw

M equivale a que no se registro el tipo de escala por lo tanto, será equivalente a Mw.

Mlv es el componente vertical de la escala Ml, por lo tanto se tratará como Ml.

Las formulas correspondeen a los criterios de equivalencia para poder estandarizar los valores de la magnitud en la escala de magnitud de momento Mw



````python
def estandarizar_magnitud_mw(df):
    df = df.copy()

    # Inicializar la nueva columna MAGNITUD_MW con los valores existentes.
    # Las magnitudes ya en Mw, M, Mww, o con unidad NaN (asumiendo que son Mw por defecto)
    # mantendrán su valor original.
    df['MAGNITUD_MW'] = df['MAGNITUD_VALOR']

    # Manejar 'Mlv' como 'Ml'
    # Esto asegura que la conversión de 'Ml' también se aplique a 'Mlv'
    # Asegurarse de que 'MAGNITUD_UNIDAD' sea de tipo string para operar
    df['MAGNITUD_UNIDAD'] = df['MAGNITUD_UNIDAD'].astype(str)
    mask_mlv = df['MAGNITUD_UNIDAD'] == 'Mlv'
    if mask_mlv.any():
        df.loc[mask_mlv, 'MAGNITUD_UNIDAD'] = 'Ml'
        print(f"✓ Se han recategorizado {mask_mlv.sum()} magnitudes de 'Mlv' a 'Ml'.")

    # Conversión para Ml (y Mlv)
    mask_ml = df['MAGNITUD_UNIDAD'] == 'Ml'
    if mask_ml.any():
        df.loc[mask_ml, 'MAGNITUD_MW'] = 0.85 * df.loc[mask_ml, 'MAGNITUD_VALOR'] + 0.53
        print(f"✓ Se han convertido {mask_ml.sum()} magnitudes de 'Ml' a 'Mw'.")

    # Conversión para Mb
    mask_mb = df['MAGNITUD_UNIDAD'] == 'Mb'
    if mask_mb.any():
        df.loc[mask_mb, 'MAGNITUD_MW'] = 0.85 * df.loc[mask_mb, 'MAGNITUD_VALOR'] + 1.03
        print(f"✓ Se han convertido {mask_mb.sum()} magnitudes de 'Mb' a 'Mw'.")

    # Conversión para Mc
    mask_mc = df['MAGNITUD_UNIDAD'] == 'Mc'
    if mask_mc.any():
        df.loc[mask_mc, 'MAGNITUD_MW'] = 0.85 * df.loc[mask_mc, 'MAGNITUD_VALOR'] + 0.53
        print(f"✓ Se han convertido {mask_mc.sum()} magnitudes de 'Mc' a 'Mw'.")

    # Conversión para Ms
    mask_ms = df['MAGNITUD_UNIDAD'] == 'Ms'
    if mask_ms.any():
        df.loc[mask_ms, 'MAGNITUD_MW'] = 0.67 * df.loc[mask_ms, 'MAGNITUD_VALOR'] + 2.07
        print(f"✓ Se han convertido {mask_ms.sum()} magnitudes de 'Ms' a 'Mw'.")

    # Las magnitudes 'M', 'Mww' y 'nan' (que pueden representar 'M' o desconocida) ya se inicializaron
    # con MAGNITUD_VALOR, por lo que no requieren un paso de conversión adicional aquí.

    # Opcional: Actualizar MAGNITUD_UNIDAD a 'Mw' para las convertidas si se desea mantener un registro
    # Para simplificar y mantener la originalidad de la unidad original, no lo haremos por defecto,
    # pero se podría añadir:
    #converted_masks = mask_ml | mask_mb | mask_mc | mask_ms
    #df.loc[converted_masks, 'MAGNITUD_UNIDAD_ESTANDAR'] = 'Mw'
    #df.loc[~converted_masks, 'MAGNITUD_UNIDAD_ESTANDAR'] = df['MAGNITUD_UNIDAD']

    print("\nPrimeras filas con la nueva columna MAGNITUD_MW:")
    print(df[['MAGNITUD_VALOR', 'MAGNITUD_UNIDAD', 'MAGNITUD_MW']].head())
    return df
````

###**8.8) Eliminar registros sismicos que no hayan ocurrido en Chile**

````python
def detectar_eventos_fuera_de_chile(df, lat_min=-56.0, lat_max=-17.0, lon_min=-90.0, lon_max=-66.0, n_ejemplos=10):
    """
    Detecta registros sísmicos fuera del rango geográfico de Chile y devuelve
    un DataFrame con los eventos que SÍ ocurrieron dentro de Chile.

    Parámetros
    ----------
    df : pd.DataFrame
        DataFrame con columnas 'LATITUD', 'LONGITUD', etc.
    lat_min : float
        Latitud mínima del bounding box (default: -56.0).
    lat_max : float
        Latitud máxima del bounding box (default: -17.0).
    lon_min : float
        Longitud mínima del bounding box (default: -90.0).
    lon_max : float
        Longitud máxima del bounding box (default: -66.0).
    n_ejemplos : int
        (No usado en la versión actual, ya que se eliminan los fuera de Chile).

    Retorna
    -------
    pd.DataFrame
        Subconjunto de df con los registros DENTRO del rango geográfico de Chile.
    """
    # Máscara para identificar eventos FUERA del rango chileno
    mascara_fuera = (
        (df['LATITUD'] < lat_min) | (df['LATITUD'] > lat_max) |
        (df['LONGITUD'] < lon_min) | (df['LONGITUD'] > lon_max)
    )

    # Copiar dataframe
    df_copia = df.copy()

    # DataFrame con eventos DENTRO del rango chileno
    df_en_chile = df_copia[~mascara_fuera]

    n_total = len(df_copia)
    n_fuera = mascara_fuera.sum() # Suma los True de la máscara
    n_en_chile = len(df_en_chile)
    porcentaje_fuera = (n_fuera / n_total) * 100

    print(f"Total de registros iniciales: {n_total}")
    print(f"Registros fuera del rango chileno (eliminados): {n_fuera} ({porcentaje_fuera:.2f}%)")
    print(f"Registros dentro del rango chileno (conservados): {n_en_chile}\n")

    # Eliminar la columna 'UBICACION' si existe y no se necesita, ahora aplicada al df filtrado
    df_en_chile = df_en_chile.drop(columns=['UBICACION'], errors='ignore')

    return df_en_chile
````

Pipeline

````python
def ejecutar_pipeline(ruta_archivo):

    print("--- Iniciando pipeline de procesamiento---")

    # Carga de datos inicial
    df = pd.read_csv(ruta_archivo)
    print(f'\n✓ Dataset cargado: {df.shape[0]:,} registros × {df.shape[1]} columnas')
    explorar_datos(df) # Llamar a la función de exploración para el estado inicial

    # 8.1) Renombrar variables del dataset a español
    df = renombrar_variables(df)
    print("\n--- 8.1 Renombrado de variables completado ---")

    # 8.2) Limpieza de datos (eliminar duplicados)
    df = limpiar_duplicados(df)
    print("--- 8.2 Limpieza de duplicados completado ---")

    # 8.3) Conversión de columnas de fecha a formato datetime de variable Fecha UTC
    df = procesar_fecha_utc(df)
    print("--- 8.3 Procesamiento de fecha UTC completado ---")

    # 8.4) Separar columna de profundidad
    df = separar_profundidad(df)
    print("--- 8.4 Separación de profundidad completado ---")

    # 8.5) Separar columna de locación (prerequisito para filtrado geográfico)
    df = separar_ubicacion(df)
    print("--- 8.5 Separación de ubicación completado ---")

    # 8.6) Separar columna de magnitud, en magnitud y unidad de medición.
    df = separar_magnitud(df)
    print("--- 8.6 Separar columna de magnitud, en magnitud y unidad de medición---")

    # 8.7) transformación de valores de las diferentes escalas sismicas.
    df = estandarizar_magnitud_mw(df)
    print("--- 8.7 transformación de valores de las diferentes escalas sismicas. ---")

    # 8.8) Eliminar registros sísmicos que no hayan ocurrido en Chile (Geographical Filtering)
    df = detectar_eventos_fuera_de_chile(df)
    print("--- 8.8) Eliminar registros sísmicos que no hayan ocurrido en Chile (Geographical Filtering) ---")

    print("--- Pipeline de procesamiento hasta la Opción 8.8 completado ---")
    return df

# Ejecutar el pipeline con la ruta global 'ruta'
df_procesado = ejecutar_pipeline(ruta)

print("\nDataFrame resultante después del pipeline (primeras 5 filas):")
print(df_procesado.head())
print(f"\nDimensiones finales del DataFrame: {df_procesado.shape[0]:,} registros × {df_procesado.shape[1]} columnas")
````

      - Numericas restantes: imputacion por mediana.

````python

def manejar_na(df):

   ¿

    df = df.copy()
    antes = len(df)

    # Variable objetivo: eliminar NA y magnitudes implausibles
    df = df.dropna(subset=['MAGNITUD'])
    df = df[df['MAGNITUD'].between(MAG_MIN, MAG_MAX)]

    # Marca temporal indispensable
    if 'FECHA_UTC' in df.columns:
        df = df.dropna(subset=['FECHA_UTC'])

    # Profundidad: imputacion por mediana
    if 'PROFUNDIDAD' in df.columns:
        df['PROFUNDIDAD'] = df['PROFUNDIDAD'].fillna(df['PROFUNDIDAD'].median())
        df['PROFUNDIDAD'] = df['PROFUNDIDAD'].clip(lower=0, upper=DEPTH_MAX)

    # Barrido final: garantiza 0 NA. Usa los verificadores de tipo de pandas
    # (robustos frente a StringDtype y datetime).
    for col in df.columns:
        if not df[col].isna().any():
            continue
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].median())
        elif pd.api.types.is_datetime64_any_dtype(df[col]):
            df = df.dropna(subset=[col])
        else:
            df[col] = df[col].fillna('Sin referencia')

    df = df.reset_index(drop=True)
    print(f'manejar_na: {antes:,} -> {len(df):,} filas (-{antes - len(df):,})')
    return df
````

###**8.8) Validación de datos**

Se validarán el procesamiento que se hizo en la etapa de limpieza y tratamiento de datos.

````python
def validar_limpieza_tratamiento(df):
    """
    Valida la limpieza y el tratamiento de los datos en el DataFrame.
    Verifica valores nulos, tipos de datos y rangos lógicos.

    Args:
        df (pd.DataFrame): El dataframe a validar después de la limpieza.

    Returns:
        bool: True si la validación es exitosa, False en caso contrario.
    """
    print("\n" + "=" * 60)
    print("INICIANDO VALIDACIÓN DE LIMPIEZA Y TRATAMIENTO DE DATOS")
    print("=" * 60)
    valid = True

    # --- 1. Verificar valores nulos en columnas críticas ---
    critical_cols_no_nan = ['FECHA_UTC', 'MAGNITUD_VALOR', 'PROFUNDIDAD_VALOR',
                            'LATITUD', 'LONGITUD', 'ANIO', 'MES', 'DIA', 'HORA']

    for col in critical_cols_no_nan:
        if col in df.columns and df[col].isnull().any():
            print(f"[FALLO] Columna '{col}' contiene valores nulos. Cantidad: {df[col].isnull().sum()}")
            valid = False
        elif col not in df.columns:
            print(f"[ADVERTENCIA] Columna crítica '{col}' no encontrada en el DataFrame.")

    # --- 2. Verificar tipos de datos ---
    expected_numeric_cols = {
        'MAGNITUD_VALOR': (float, int),
        'PROFUNDIDAD_VALOR': (float, int),
        'LATITUD': float,
        'LONGITUD': float,
        'ANIO': int,
        'MES': int,
        'DIA': int,
        'HORA': int,
        'DISTANCIA_EPICENTRO_KM': (float, int)
    }
    for col, expected_type in expected_numeric_cols.items():
        if col in df.columns:
            if not pd.api.types.is_numeric_dtype(df[col]):
                print(f"[FALLO] Columna '{col}' no es de tipo numérico. Tipo actual: {df[col].dtype}")
                valid = False
            elif isinstance(expected_type, tuple) and not any(pd.api.types.is_dtype(df[col], t) for t in expected_type):
                 pass # Allow int or float for numeric cols
            elif not isinstance(expected_type, tuple) and not pd.api.types.is_dtype(df[col], expected_type):
                print(f"[FALLO] Columna '{col}' no es del tipo '{expected_type}'. Tipo actual: {df[col].dtype}")
                valid = False

    if 'FECHA_UTC' in df.columns and not pd.api.types.is_datetime64_any_dtype(df['FECHA_UTC']):
        print(f"[FALLO] Columna 'FECHA_UTC' no es de tipo datetime. Tipo actual: {df['FECHA_UTC'].dtype}")
        valid = False

    # --- 3. Verificar rangos lógicos ---
    # Se asume que las constantes LAT_MIN, LAT_MAX, LON_MIN, LON_MAX, MAG_MIN, MAG_MAX, DEPTH_MAX están definidas
    # Si no, deberían pasarse como argumentos o definirse globalmente antes de llamar a esta función.
    try:
        global LAT_MIN, LAT_MAX, LON_MIN, LON_MAX, MAG_MIN, MAG_MAX, DEPTH_MAX
    except NameError:
        print("[ADVERTENCIA] Las variables de límites (LAT_MIN, etc.) no están definidas. Omitiendo validación de rangos.")
        # Definir valores por defecto si no están en el contexto
        LAT_MIN, LAT_MAX = -60.0, -15.0
        LON_MIN, LON_MAX = -80.0, -64.0
        MAG_MIN, MAG_MAX = -3.0, 12.0
        DEPTH_MAX        = 700.0

    if 'LATITUD' in df.columns and not df['LATITUD'].between(LAT_MIN, LAT_MAX).all():
        print(f"[FALLO] 'LATITUD' fuera de rango ({LAT_MIN} a {LAT_MAX}).")
        valid = False
    if 'LONGITUD' in df.columns and not df['LONGITUD'].between(LON_MIN, LON_MAX).all():
        print(f"[FALLO] 'LONGITUD' fuera de rango ({LON_MIN} a {LON_MAX}).")
        valid = False
    if 'MAGNITUD_VALOR' in df.columns and not df['MAGNITUD_VALOR'].between(MAG_MIN, MAG_MAX).all():
        print(f"[FALLO] 'MAGNITUD_VALOR' fuera de rango ({MAG_MIN} a {MAG_MAX}).")
        valid = False
    if 'PROFUNDIDAD_VALOR' in df.columns and not df['PROFUNDIDAD_VALOR'].between(0, DEPTH_MAX).all():
        print(f"[FALLO] 'PROFUNDIDAD_VALOR' fuera de rango (0 a {DEPTH_MAX}).")
        valid = False
    if 'ANIO' in df.columns and not df['ANIO'].between(2000, 2024).all(): # Rango específico del dataset
        print("[FALLO] 'ANIO' fuera de rango (2000 a 2024).")
        valid = False

    # --- 4. Verificar consistencia en unidades ---
    if 'PROFUNDIDAD_UNIDAD' in df.columns and not all(df['PROFUNDIDAD_UNIDAD'].isin(['km', 'desconocido'])):
        print("[ADVERTENCIA] 'PROFUNDIDAD_UNIDAD' contiene valores inesperados aparte de 'km' o 'desconocido'.")

    if 'MAGNITUD_UNIDAD' in df.columns and 'desconocido' in df['MAGNITUD_UNIDAD'].unique():
        print("[ADVERTENCIA] 'MAGNITUD_UNIDAD' contiene 'desconocido'. Considerar normalización.")

    # --- 5. Resultados finales ---
    if valid:
        print("\n[ÉXITO] La validación de la limpieza y tratamiento de datos fue exitosa.")
    else:
        print("\n[FALLO] Se encontraron problemas durante la validación de la limpieza y tratamiento de datos.")

    return valid

# Ejemplo de uso (descomentar para probar)
# if 'df1' in locals():
#     estado_limpieza = validar_limpieza_tratamiento(df1)
# else:
#     print("Error: df1 no está definido. Asegúrate de ejecutar los pasos anteriores.")
````
